```python
'''Given: A quality threshold value q, percentage of bases p

, and set of FASTQ entries.

Return: Number of reads in filtered FASTQ entries'''

from Bio import SeqIO
```

```python
with open('rosalind_filt.txt','r+') as fp:
    # read an store all lines into list
    lines = fp.readlines()
    # move file pointer to the beginning of a file
    fp.seek(0)
    # truncate the file
    fp.truncate()
    threshold = int(lines[0][:2])
    prob = int(lines[0][3:5])
    # start writing lines except the first line
    # lines[1:] from line 2 to last line
    fp.writelines(lines[1:])
    
```

```python
sum = 0
counter =0
for record in SeqIO.parse('rosalind_filt.txt',"fastq"):
    sum=0
    for phrd in record.letter_annotations["phred_quality"]:
        if phrd >=threshold:
            sum +=1
    if (sum/len(record.letter_annotations["phred_quality"]))*100 >=prob:
        counter+=1
    
```
